```python
#Create a Line Chart Using the MATLAB Method

%matplotlib inline
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create a Line Chart Using the Object-oriented interface
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plo and adjust the color, width and marker
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Using the object-oriented approach

#Edit line structure
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label="Boston")
# Set x label
ax.set_xlabel("Months")
# Set y label
ax.set_ylabel("Fares($)")
# Set Title
ax.set_title("PyBer Fare by Month")
# Set y-axis limit
ax.set_ylim(0, 50)
# Add grid lines to chart
ax.grid()
```

```python
# Creating Bubble Charts
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# To increase the size of the markers
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# second method
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Create a Bubble Chart Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
#Skill Drill: change markers to sky blue, 5x the size, color 20% transparent, add black edge color to the circles, make linewidth of the circles 2 pts
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s = [i * 5 for i in y_axis], c="skyblue", alpha=0.2, edgecolors='black',linewidths=2)
# Set title
ax.set_title("PyBer Fare by Month")
#Set x and y labels
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")
# Add a limit for x-axis
ax.set_xlim(0, 50)
# Add a legend for city of Boston
ax.legend("Boston")
# Invert y-axis so that Jan is at the top
ax.invert_yaxis()
```

```python
# Create a Pie Chart Using the MATLAB Method
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
#increase the size of the pie-chart
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, explode=explode_values, colors=colors, labels=x_axis, autopct='%.1f%%')
plt.show()
```
